## Parser

#### CFG

Context-Free Grammar (CFG)

S  HTML

HTML  <\html> HEAD BODY <\/html>

HEAD  <\head> TEXT <\/head> | 

BODY  <\body> TEXT <\/body> | 

TEXT  <\title> TEXT <\/title> TEXT | <\h1> TEXT <\/h1> TEXT | <\p> TEXT <\/p> TEXT | <\button> TEXT <\/button> TEXT | 

---

#### Non-Terminals and Terminals

| Non-Terminals | Terminals                                                                                           |
|---------------|-----------------------------------------------------------------------------------------------------|
| S             | `<html>`, `</html>`, `<head>`, `</head>`, `<body>`, `</body>`, `<title>`, `</title>`, `<h1>`, `</h1>`, `<p>`, `</p>`, `<button>`, `</button>`, `EOS` |

---

Kolom selanjutnya



---

#### Notes
- `epsilon` represents an empty production.
- The default entry for unspecified transitions is `['error']`.

```python
tokens = sentence.lower().split()
tokens.append('EOS')
```

```python
# Definisi simbol
non_terminals = ['S', 'HTML', 'HEAD', 'BODY', 'TEXT']
terminals = ['<html>', '</html>', '<head>', '</head>', '<body>', '</body>', '<title>', '</title>', '<h1>', '</h1>', '<p>', '</p>', '<button>', '</button>']
```

```python
# Inisialisasi Parse Table
parse_table = {}

# Definisi Parse Table HTML
parse_table[('S', '<html>')] = ['HTML']
parse_table[('HTML', '<html>')] = ['<html>', 'HEAD', 'BODY', '</html>']
parse_table[('HEAD', '<head>')] = ['<head>', 'TEXT', '</head>']
parse_table[('HEAD', '<body>')] = ['epsilon']  # No HEAD content
parse_table[('HEAD', '</html>')] = ['epsilon']  # Directly to </html>

parse_table[('BODY', '<body>')] = ['<body>', 'TEXT', '</body>']
parse_table[('BODY', '</html>')] = ['epsilon']  # No BODY content

parse_table[('TEXT', '<title>')] = ['<title>', 'TEXT', '</title>', 'TEXT']
parse_table[('TEXT', '<h1>')] = ['<h1>', 'TEXT', '</h1>', 'TEXT']
parse_table[('TEXT', '<p>')] = ['<p>', 'TEXT', '</p>', 'TEXT']
parse_table[('TEXT', '<button>')] = ['<button>', 'TEXT', '</button>', 'TEXT']

# Handling tag penutup
parse_table[('TEXT', '</title>')] = ['epsilon']
parse_table[('TEXT', '</h1>')] = ['epsilon']
parse_table[('TEXT', '</p>')] = ['epsilon']
parse_table[('TEXT', '</button>')] = ['epsilon']
parse_table[('TEXT', '</head>')] = ['epsilon']
parse_table[('TEXT', '</body>')] = ['epsilon']
parse_table[('TEXT', '</html>')] = ['epsilon']

# Jika Transisi Error
for nt in non_terminals:
    for t in terminals:
        if (nt, t) not in parse_table:
            parse_table[(nt, t)] = ['error']
```

```python
def read_html_file(file_path):
    with open(file_path, 'r') as file:
        return file.read()
```

```python
import re

def tokenize_html(html_content):
    # Regular expression untuk mencocokkan Tag HTML
    tokens = re.findall(r'</?[^>]+>', html_content)
    tokens.append('EOS')  # Append End Of Stream token
    return tokens
```

```python
def parse(tokens, sentence):
    stack = ['#', 'S']
    idx_token = 0
    symbol = tokens[idx_token]

    while len(stack) > 0:
        top = stack[-1]
        print('top =', top)
        print('symbol =', symbol)
        if top in terminals:
            print('top is a terminal symbol')
            if top == symbol:
                stack.pop()
                idx_token += 1
                if idx_token < len(tokens):
                    symbol = tokens[idx_token]
                else:
                    symbol = 'EOS'
                if symbol == 'EOS':
                    print('stack content:', stack)
                    stack.pop()
            else:
                print('error: terminal mismatch')
                break
        elif top in non_terminals:
            print('top is a non-terminal symbol')
            if (top, symbol) in parse_table:
                rule = parse_table[(top, symbol)]
                if len(rule) > 0 and rule[0] != 'error':
                    stack.pop()
                    if rule[0] != 'epsilon':
                        for r in reversed(rule):
                            stack.append(r)
                elif len(rule) == 0 or rule[0] == 'epsilon':
                    stack.pop()
                else:
                    print('error: encountered error rule')
                    break
            else:
                print('error: no rule for this non-terminal and terminal')
                break
        else:
            # Skip untuk non-tag content (text content)
            if re.match(r'<[^>]+>', symbol) is None:
                idx_token += 1
                if idx_token < len(tokens):
                    symbol = tokens[idx_token]
                else:
                    symbol = 'EOS'
            else:
                print('error: unknown symbol')
                break
        print('stack content:', stack)
        print()
    
    # Kesimpulan
    print()
    if symbol == 'EOS' and len(stack) == 0:
        print('Input string:', sentence, 'diterima, sesuai Grammar')
    else:
        print('Error, input', sentence, ', tidak diterima, tidak sesuai grammar')
```

```python
# Contoh Implementasi
html_content = read_html_file('input1.html')
tokens = tokenize_html(html_content)
parse(tokens, html_content)
```
